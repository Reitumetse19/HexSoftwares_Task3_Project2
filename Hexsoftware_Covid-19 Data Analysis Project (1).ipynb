### Average Weekly % Increase in Confirmed COVID-19 Cases by WHO Region

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style = 'whitegrid')

region_growth = covid19_df.groupby('WHO Region')['1 week % increase'].mean().sort_values(ascending =False)

plt.figure(figsize = (8, 4))
sns.barplot(x = region_growth.values, y = region_growth.index, palette = 'rainbow')
plt.title('Average Weekly % Increase in Confirmed COVID-19 Cases by WHO Region')
plt.xlabel('Average 1 Week % Increase')
plt.ylabel('WHO Region')
plt.tight_layout()
plt.show()
```

### Weekly % Increase in Confirmed Cases in South-East Asia

```python
South_East_Asia_df = covid19_df[covid19_df['WHO Region'] == 'South-East Asia']

Growth_SEA = South_East_Asia_df.sort_values(by = '1 week % increase', ascending = False)

plt.figure(figsize = (10, 4))
sns.barplot(x = '1 week % increase', y = 'Country/Region', data = Growth_SEA, palette = 'spring')
plt.title('Weekly % Increase in Confirmed Cases in South-East Asia')
plt.xlabel('1 Week % Increase')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Comparison of Death and Recovery Rates in South-East Asia

```python
import matplotlib.pyplot as plt
import seaborn as sns

South_East_Asia_df = covid19_df[covid19_df['WHO Region'] == 'South-East Asia']

South_East_Asia = South_East_Asia_df.sort_values(by ='Deaths / 100 Cases', ascending = False)

plt.figure(figsize = (6, 4))
sns.set(style = 'whitegrid')

bar_width = 0.60
x = range(len(South_East_Asia))

plt.bar(x, South_East_Asia['Deaths / 100 Cases'], width = bar_width, label ='Deaths / 100 Cases', color = 'red')
plt.bar([i + bar_width for i in x], South_East_Asia['Recovered / 100 Cases'], width = bar_width, label = 'Recovered / 100 Cases', color = 'yellow')

plt.xticks([i + bar_width / 2 for i in x], South_East_Asia['Country/Region'], rotation=90)
plt.ylabel('Rate per 100 Cases')
plt.title('Comparison of Death and Recovery Rates in South-East Asia')
plt.legend()
plt.tight_layout()
plt.show()


```

### Proportion of Total COVID-19 Deaths by WHO Region

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Regional_deaths = covid19_df.groupby('WHO Region')['Deaths'].sum().sort_values(ascending = False)

plt.figure(figsize = (7, 7))
plt.pie(Regional_deaths, labels = Regional_deaths.index, autopct="%1.1f%%", startangle = 180, colors = sns.color_palette('rainbow'))
plt.title('Proportion of Total COVID-19 Deaths by WHO Region')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

### Confirmed vs Recovered COVID-19 Cases of Top 15 Countries

```python
import pandas as pd
import matplotlib.pyplot as plt

confirmed_vs_recovered =  covid19_df.sort_values(by = 'Confirmed', ascending = False).head(15)

plt.figure(figsize=(10, 4))

plt.plot(confirmed_vs_recovered['Country/Region'], confirmed_vs_recovered['Confirmed'], marker ='o', label ='Confirmed Cases', color = 'm')

plt.plot(confirmed_vs_recovered['Country/Region'], confirmed_vs_recovered['Recovered'], marker ='o', label ="Recovered Cases", color='red')

plt.xticks(rotation=45)
plt.ylabel('Number of Cases')
plt.title('Confirmed vs Recovered COVID-19 Cases of Top 15 Countries')
plt.legend()
plt.tight_layout()
plt.show()

```

### Correlation Between Active Cases and Death Rate

```python
import pandas as pd

Crr = covid19_df[['Country/Region', 'Active', 'Deaths / 100 Cases']].dropna()

Correlation = Crr["Active"].corr(Crr['Deaths / 100 Cases'])
print(f"Pearson correlation: {Correlation:.3f}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (8, 4))
sns.scatterplot(x ='Active', y = 'Deaths / 100 Cases', data = Crr)

plt.title('Correlation Between Active Cases and Death Rate')
plt.xlabel('Active Cases')
plt.ylabel('Deaths / 100 Cases')
plt.tight_layout()
plt.show()
```
